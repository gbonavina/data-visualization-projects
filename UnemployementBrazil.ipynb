```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("unemployment analysis.csv")
df.head(n=25)
```

```python
country_name = []
country_code = [] 
year = []
unemployment_rate = []
column_names = df.columns[2:]

for index, row in df.iterrows():
    for temp_year in column_names:
        country_name.append(row['Country Name'])
        country_code.append(row['Country Code'])
        year.append(temp_year)
        unemployment_rate.append(row[temp_year])
        
df2 = pd.DataFrame(data={'Year': year,'Country Name': country_name,'Country Code': country_code,'Unemployement Rate': unemployment_rate})
df2.head()
```

```python
br_df = df2[744:775]
display(br_df)
```

```python
final_df = br_df[["Year", "Unemployement Rate"]]
display(final_df)
```

```python
plt.figure(figsize=(20,4))
plt.plot(final_df['Year'], final_df['Unemployement Rate'], label='Unemployement Rate')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate (%)')
plt.title('Unemployement Rate in Brazil from 1991 to 2021')
plt.legend()
plt.show()
```
